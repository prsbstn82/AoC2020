```python
import os
import numpy as np
import pandas as pd
```

```python
data = []
with open("input.csv", "r") as file:
    for row in file:
        data.append(row[:-1])
```

```python
nb_tree = 0
x = 0
y = 0
len_row = len(data[0])
while x < len(data)-1:
    x+=1
    y+=1
    y_bis = y % len_row
    if data[x][y_bis] == "#":
        nb_tree +=1
    else:
        pass
```
