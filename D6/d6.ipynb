```python
data = []
with open("input.csv", "r") as file:
    for row in file:
        data.append(row)
```

```python
data2 = []
gp = []
i = 0
for i in range(len(data)):
    if data[i] == "\n":
        data2.append(gp)
        gp=[]
    else:
        gp.append(data[i][:-1])

data2.append(['qxvuhyp', 'hxvcpy'])
```

```python
count2 = []
for gp in data2:
    l = []
    for rep in gp:
        l.append([char for char in rep])
    if len(l) == 1:
        count2.append(len(np.unique(l[0])))
    elif len(l) == 2:
        unique = list(set(l[0]).intersection(l[1]))
        count2.append(len(unique))
    else:
        unique = list(set(l[0]).intersection(l[1]))
        for i in range(2, len(l)):
            unique = list(set(unique).intersection(l[i]))
        count2.append(len(unique))
```
