```python
import os
import numpy as np
import pandas as pd
import re
```

```python
data = []
with open("input.txt", "r") as file:
    for row in file:
        data.append(row)
```

```python
i=0
passeports = []
while i < len(data)-1:
    pssprt = []
    while data[i] != "\n":
        pssprt.append(data[i])
        i+=1
    passeports.append(pssprt)
    i+=1
    
passeports.append([data[i]])
```

```python
data2 = []
for p in passeports:
    pp = {}
    for info in p:
        liste_info = info.split("\n")[0].split(" ")
        for ss_info in liste_info:
            pp[ss_info.split(":")[0]] = ss_info.split(":")[1]
    data2.append(pp)
```

```python
df = pd.DataFrame(data2)
```

```python
def is_valid(eyr, hcl, hgt, byr, iyr, pid, ecl):
    pattern_hcl = re.compile("^#([a-fA-F0-9]{6}|[a-fA-F0-9]{3})$")
    pattern_pid = re.compile("^\d{9}$")
    eyr_ok = False
    hcl_ok = False
    hgt_ok = False
    byr_ok = False
    iyr_ok = False
    pid_ok = False
    ecl_ok = False
    
    if byr == np.nan or type(byr) == float:
        pass
    elif (len(byr) == 4) and (int(byr) >= 1920) and (int(byr) <= 2002):
        byr_ok = True
    else:
        pass
    
    if iyr == np.nan or type(iyr) == float:
        pass
    elif (len(iyr) == 4) and (int(iyr) >= 2010) and (int(iyr) <= 2020):
        iyr_ok = True
    else:
        pass
    
    if eyr == np.nan or type(eyr) == float:
        pass
    elif (len(eyr) == 4) and (int(eyr) >= 2020) and (int(eyr) <= 2030):
        eyr_ok = True
    else:
        pass
    
    if hgt == np.nan or type(hgt) == float:
        pass
    elif hgt[-2:] in ["cm", "in"]:
        if hgt[-2:] == "cm" and int(hgt[:-2]) >= 150 and int(hgt[:-2]) <= 193:
            hgt_ok = True
        elif hgt[-2:] == "in" and int(hgt[:-2]) >= 59 and int(hgt[:-2]) <= 76:
            hgt_ok = True
        else:
            pass
    else:
        pass
    if hcl == np.nan or type(hcl) == float:
        pass
    elif pattern_hcl.match(hcl):
        hcl_ok = True
    else:
        pass
    
    if ecl == np.nan:
        pass
    elif ecl in ["amb", "blu", "brn", "gry", "grn", "hzl", "oth"]:
        ecl_ok = True
        pass
    else:
        pass
    
    if pid == np.nan or type(pid) == float:
        pass
    elif pattern_pid.match(pid):
        pid_ok = True
        pass
    else:
        pass
    
    print(f"""
    byr:: {byr}: {byr_ok}
    iyr:: {iyr}: {iyr_ok}
    eyr:: {eyr}: {eyr_ok}
    hgt:: {hgt}: {hgt_ok}
    hcl:: {hcl}: {hcl_ok}
    ecl:: {ecl}: {ecl_ok}
    pid:: {pid}: {pid_ok}
    {eyr_ok and hcl_ok and hgt_ok and byr_ok and iyr_ok and pid_ok and ecl_ok}
    """)
    
    return eyr_ok and hcl_ok and hgt_ok and byr_ok and iyr_ok and pid_ok and ecl_ok
        
```

```python
df["is_valid2"] = df.apply(lambda row: is_valid(row["eyr"], row["hcl"], row["hgt"], row["byr"], row["iyr"], row["pid"], row["ecl"]), axis=1)
```
